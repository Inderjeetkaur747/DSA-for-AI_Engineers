```python
# Problem 1 – Find Maximum Element
# Given an array, return maximum element.
def find_max(arr):
  if not arr:
    return None
  max_ele = arr[0]
  for num in range(1, len(arr)):
    if arr[num] > max_ele:
      max_ele = arr[num]
  return max_ele

arr = [1,2,4,3,6,9,8]
print(find_max(arr))
```

```python
# Problem 3 – Move Zeros
# Given an integer array nums, move all 0's to the end
# while maintaining the relative order of non-zero elements.
# Do it in-place.

def move_zeros(arr):
  insert_arr = 0
  for num in range(len(arr)):
    if arr[num]!=0:
      arr[insert_arr] = arr[num]
      insert_arr+=1

  for num in range(insert_arr, len(arr)):
    arr[num]=0
  return arr

arr = [1,3,0,4,0,2,0,5]
print(move_zeros(arr))
```

```python
# Problem 4 – Remove Element (LeetCode 27)
# Given array nums and a value val,
# remove all occurrences of val in-place.
# Return the new length.
def remove_element(arr,val):
  write_index = 0
  for i in range(len(arr)):
    if arr[i]!=val:
      arr[write_index] = arr[i]
      write_index+=1
  return write_index

arr = [1,2,2,3,4]
val = 2
print(remove_element(arr,val))
```

```python
# Problem 5 – Best Time to Buy and Sell Stock (LeetCode 121)
# Given:
# prices = [7,1,5,3,6,4]
# Return maximum profit.
# Rules:
# You must buy before selling.
# Only one transaction allowed.
# Example:
# Buy at 1
# Sell at 6
# Profit = 5
def max_profit(prices):
  if not prices:
    return 0

  min_price = prices[0]
  max_profit = 0
  for i in range(1, len(prices)):
    if prices[i] < min_price:
      min_price = prices[i]
    else:
      profit = prices[i] - min_price
      max_profit = max(max_profit,profit)

  return max_profit


prices = [7,1,5,3,6,4]
print(max_profit(prices))


```

```python
# Problem 6 – Contains Duplicate (LeetCode 217)
# Given an array, return True if any value appears at least twice.
# Example:
# [1,2,3,1] → True
# [1,2,3,4] → False
# ⚠ Important:
# Do NOT use sorting.
# Do NOT use built-in set.
# Think pure array logic first.
def contains_duplicate(val):
  n = len(val)
  for i in range(n):
    for j in range(i+1, n):
      if val[i]==val[j]:
        return True

  return False


val = [1,2,3,1]
print(contains_duplicate(val))

# Outer loop = n
# Inner loop = n

# Total = n²

# Time Complexity = O(n²)
# Space Complexity = O(1)
# This problem shows:
# Arrays alone are inefficient for lookup problems.
# We need faster lookup.
# And that’s exactly why Hashing exists.
#Arrays are bad at searching quickly.
```

```python
def contains_duplicate(arr):
  seen = set()
  for num in arr:
    if num in seen:
      return True
    seen.add(num)
  return False

arr = [1,2,3,1]
print(contains_duplicate(arr))
```

```python
# Problem – Two Sum (LeetCode 1)
# Given:
# nums = [2,7,11,15]
# target = 9
# Return indices of two numbers that add to target.
def Two_sum(num, target):
  hash_map = {}
  for i, n in enumerate(num):
    complement = target - n
    if complement in hash_map:
      return [hash_map[complement],i]
    hash_map[n] = i
  return []

num = [2,7,11,15]
target = 9
print(Two_sum(num, target))

```
